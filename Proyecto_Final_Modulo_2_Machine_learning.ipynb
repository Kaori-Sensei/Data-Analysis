**Perceptrón**. El perceptrón fue formulado por Frank Rosenblatt entre los años 1950 y 1960, inspirado en los trabajos de Warren McCulloch y Walter Pitts. Hoy en día se utiliza modelos de mayor alcance que fueron desarrollados en tiempos más recientes, como el de neurona *sigmoidea* que se fundamenta en la función lógistica

\begin{equation}
f(x)=\frac{1}{1-e^{-x}}.
\end{equation}

El *percetrón*, ofrece características básicas semejantes a lo que expresa la función anterior, y refleja de manera explicita el funcionamiento básico que subyace a nuestra concepción cómun de neurona.

La idea básica por detrás de las neuronas artificiales es trabajar como funciones que retornan una salida a partir de un conjunto de entradas, cada una de las cuales actúa con una influencia diferente que se determina mediante un conjunto de pesos. Al pretender asemejar la función neuronal, se implementa utilizando como salida un espectro binario.

En el percetrón original la salida es determinada por la suma $\sum_j w_jx_j$, dependiendo de si su valor es mayor o menor que algún valor de referencia. Es decir,
\begin{equation}
\text{Salida}=\left|
  \begin{cases}
    0 & \text{si} \quad \sum_j w_jx_j \leq b\\
    1 & \text{de otro modo}
    \end{cases}
  \right.
\end{equation}

# **Redes neuronales convolucionales**

Las redes neuronales convolucionales son un modelo en el que se pretende reducir mediante operaciones de convolución la dimensión de los datos de entrada. Para esto, luego de la capa de entrada, se implementa una capa convoluciones que resume mediante filtros los valores en los datos de entrada. Ésto se hace operando escalarmente subconjuntos de los datos de entrada con matrices que definen filtros. Los filtros se desplazan a través de los datos de entrada, y al final se seleccionan valores según sus valores máximos o mínimos. Al final, a la capa de salida le llegan conjuntos de datos de tamaño reducido de manera que facilite el discernimiento de la salida más adecuada.

## **Funciones propias para modelación**

###**Modelo secuencial**
### Definición de modelo secuencial
 Número de *hidden layers* igual al tamaño de arreglo de capas.
Modelos así son útiles cuando se tiene 1 tensor de entreda y uno de salida por capa. Se compila con *stochastic gradient descent*, *categorical cross entropy*, y *accuracy* como *score*.

```python
#========================================================================
# Modelo secuencial: definición de arquitectura
#========================================================================
# Definición de modelo secuencial hidden layers igual al tamaño de ar. capas.
# Modelos así son útiles cuando se tiene 1 tensor de entreda y uno de salida por cpa
# Se compila con stochastic gradient descent,categorical cross entropy, y accuracy.
#========================================================================
def secuencial(capas):
    model=Sequential()

    # # Capa entrada
    model.add(Dense(capas[0], activation='sigmoid', input_shape=(vector_size,)))
    # # Capas ocultas
    for s in capas[1:]:
        model.add(Dense(units = s, activation = 'sigmoid'))
    # # Capa salida
    model.add(Dense(units=nclases, activation='softmax'))
    model.compile(optimizer="sgd", loss='categorical_crossentropy', metrics=['accuracy'])
    model.summary()

    return model
```

##**Función de evaluación**

```python
def evaluate(model, x_train,y_train,x_test,y_test,fig,gind,totplots,ncolu,batch_size=128, epochs=5):

    history=model.fit(x_train,y_train,batch_size=batch_size,epochs=epochs,validation_split=.1,verbose=False)
    loss,accuracy=model.evaluate(x_test,y_test,verbose=False)
    nfi=totplots//ncolu+totplots%ncolu
    ax=fig.add_subplot(nfi,ncolu,gind)
    ax.plot(history.history['accuracy'])
    ax.plot(history.history['val_accuracy'])
    ax.legend(['Entrenamiento','Validación'], loc='best')

    print('Función de pérdida para prubeas: {:.3}\n Accuracy para pruba: {:.3}'.format(loss,accuracy))
```

## **Red convolucional**

### Modelo

```python
def convolucional():

    modelo=Sequential()
    modelo.add(layers.Conv2D(32,(3,3),activation='relu',input_shape=(28,28,1)))
    modelo.add(layers.MaxPooling2D((2,2)))
    modelo.add(layers.Conv2D(64,(3,3),activation='relu'))
    modelo.add(layers.MaxPooling2D((2,2)))
    modelo.add(layers.Conv2D(64, (3,3),activation='relu'))
    modelo.add(layers.Flatten())
    modelo.add(layers.Dense(64,activation='relu'))
    modelo.add(layers.Dense(10))
    modelo.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])
    modelo.summary()

    return modelo

```

## Evaluación

```python
#====================================================================
# Función de evaluación para modelo convolucional
#====================================================================
def evaconv(modelo,x_train,y_train,x_test,y_test,fig,batch_size=128,epochs=5):

    history=modelo.fit(x_train,y_train,batch_size=batch_size,epochs=epochs,validation_data=(x_test,y_test),verbose=2)
    loss,accuracy=modelo.evaluate(x_test,y_test,verbose=1)
    #nfi,nclou=1,1
    ax=plt.gca()
    ax.plot(history.history['accuracy'])
    ax.plot(history.history['val_accuracy'])
    ax.legend(['Entrenamiento','Validación'], loc='best')

    print('Función de pérdida para prubeas: {:.3}\n Accuracy para pruba: {:.3}'.format(loss,accuracy))
```

# **3. Preparacion de los datos**

## **Preprocesamiento de datos**
Los valores en los conjuntos de imágenes no se normalizan entre 0 y 1. Para las etiquetas se utiliza *One Hot Encoder*, puesto que no se desea respetar la relación de órden implícita en los dígitos de las imágenes por predecir  

```python
#====================================================================
# Preprocesamiento de datos
#====================================================================
def preprocess(X_train,Y_train,X_test,Y_test):

    X_train=X_train.reshape(X_train.shape[0],vector_size)
    X_test=X_test.reshape(X_test.shape[0],vector_size)
    # # One Hot Encodingg
    y_train=keras.utils.np_utils.to_categorical(Y_train,nclases)
    y_test=keras.utils.np_utils.to_categorical(Y_test,nclases)
    print("#  Resultados One Hot Encoder para: {}\n {}".format(Y_train[:5], y_train[:5]))
    return X_train,y_train,X_test,y_test
```

## **SVM - Support Vector Machine**

```python
from sklearn import datasets
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
import numpy as np

# Cargar el conjunto de datos MNIST
#digits = datasets.load_digits()
mnist = fetch_openml('mnist_784')
# Dividir los datos en conjuntos de entrenamiento y prueba
#X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(mnist.data, mnist.target, test_size=0.2, random_state=42)
# Entrenar el modelo SVM
svm_model = SVC(kernel='linear', C=1, gamma='auto')
svm_model.fit(X_train, y_train)

# Realizar predicciones sobre el conjunto de prueba
y_pred = svm_model.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f'Precisión del modelo: {accuracy}')

# Calcular la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
print(f'Matriz de confusión:\n{conf_matrix}')

# Visualizar algunos ejemplos de dígitos escritos a mano
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(8, 8))
for i, ax in enumerate(axes.flat):
    ax.imshow(np.reshape(X_test[i], (8, 8)), cmap='gray')
    ax.set_title(f'Etiqueta verdadera: {y_test[i]}\nEtiqueta predicha: {y_pred[i]}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```
