```python
# ============================================================
# AGREGAR UNA BARRA DE MEDIDA A UNA IMAGEN EN COLAB
# ============================================================

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# ============================================================
# 1. ESCRIBE AQUÍ LA RUTA EXACTA DE TU IMAGEN
#    Ejemplo: "/content/gulupa_control.png"
# ============================================================

img_path = "/content/mosca de la fruta_Gulupa (1).png"   # <-- CAMBIA ESTA RUTA

# ============================================================
# 2. VERIFICAR SI EL ARCHIVO EXISTE
# ============================================================

if not os.path.exists(img_path):
    raise FileNotFoundError(
        f"No se encontró el archivo en esta ruta:\n{img_path}\n\n"
        "Verifica que la imagen esté subida en /content/ y que el nombre sea exacto."
    )

print("✅ Imagen encontrada:", img_path)

# ============================================================
# 3. CARGAR IMAGEN
# ============================================================

img = mpimg.imread(img_path)

# ============================================================
# 4. CREAR FIGURA
# ============================================================

fig, ax = plt.subplots(figsize=(6, 6), dpi=300)
ax.imshow(img)
ax.axis("off")

# ============================================================
# 5. AJUSTAR LA POSICIÓN DE LA BARRA DE MEDIDA
#    CAMBIA ESTOS VALORES SEGÚN TU IMAGEN
# ============================================================

x_bar = 70         # posición horizontal de la barra
y_top = 80         # punto superior de la barra
y_bottom = 430     # punto inferior de la barra

measure_text = "~7.0 cm"   # cambia el valor por tu medida real

# ============================================================
# 6. DIBUJAR BARRA DE MEDIDA
# ============================================================

# Línea vertical principal
ax.plot(
    [x_bar, x_bar],
    [y_top, y_bottom],
    color="#2b6cb0",
    linewidth=2.5
)

# Línea horizontal superior
ax.plot(
    [x_bar - 10, x_bar + 10],
    [y_top, y_top],
    color="#2b6cb0",
    linewidth=2.5
)

# Línea horizontal inferior
ax.plot(
    [x_bar - 10, x_bar + 10],
    [y_bottom, y_bottom],
    color="#2b6cb0",
    linewidth=2.5
)

# Texto de la medida
ax.text(
    x_bar - 25,
    (y_top + y_bottom) / 2,
    measure_text,
    rotation=90,
    va="center",
    ha="center",
    fontsize=12,
    fontweight="bold",
    color="black"
)

# ============================================================
# 7. GUARDAR RESULTADO
# ============================================================

output_path = "/content/imagen_con_medida.png"

plt.tight_layout()
plt.savefig(output_path, dpi=300, bbox_inches="tight", pad_inches=0.05)
plt.show()

print("✅ Imagen final guardada en:", output_path)
```
